```python
from tkinter import *
from PIL import Image, ImageTk
import joblib

def show_entry_fields():
    try:
        # Extracting values from the entries and dropdowns
        p1 = int(e1.get())
        p2 = int(gender_var.get().split(' ')[0])  # Extract numeric part from "0 (Female)" or "1 (Male)"
        p3 = int(ChestPain_var.get().split('(')[0])  # Extract numeric part before the '(' character
        p4 = int(e4.get())
        p5 = int(e5.get())
        p6 = int(fastingbloodsugar_var.get())
        p7 = int(restecg_var.get())
        p8 = int(e8.get())
        p9 = int(ExerciseInducedAngina_var.get().split('(')[0])
        p10 = float(e10.get())
        p11 = float(e11.get())
        p12 = int(e12.get())
        p13 = int(Thalassemia_var.get().split('(')[0])

        # Load the model and make prediction
        model = joblib.load('model_joblib_heart')
        result = model.predict([[p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11, p12, p13]])
        
        # Debug: print result to console
        print(f"Prediction result: {result}")

        result_window = Toplevel(master)
        result_window.attributes('-fullscreen', True)
        
        if result == 0:
            result_text = "No Heart Disease"
            result_bg = "lightgreen"
            result_image_path = r"C:\jupyter notebook\WhatsApp Image 2024-07-24 at 23.32.22.jpeg"
            causes_text = "You are healthy. Continue maintaining a healthy lifestyle."
        else:
            result_text = "Possibility of Heart Disease"
            result_bg = "lightcoral"
            result_image_path = r"C:\jupyter notebook\heart-attack.webp"
            causes_text = determine_causes(p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11, p12, p13)
        
        result_window.configure(bg=result_bg)
        
        # Result text label
        result_label = Label(result_window, text=result_text, font=('Helvetica', 28, 'bold'), bg=result_bg, fg="black", relief=RAISED, padx=20, pady=20)
        result_label.pack(pady=30, padx=30)
        
        # Causes text label
        causes_label = Label(result_window, text=causes_text, font=('Helvetica', 18), bg=result_bg, fg="black", wraplength=600)
        causes_label.pack(pady=20, padx=30)
        
        # Load and display the result image
        result_image = Image.open(result_image_path)
        result_image = result_image.resize((400, 400), Image.Resampling.LANCZOS)
        result_photo = ImageTk.PhotoImage(result_image)
        
        result_image_label = Label(result_window, image=result_photo, bg=result_bg, relief=RAISED, padx=10, pady=10)
        result_image_label.image = result_photo
        result_image_label.pack(pady=20)
        
        # Exit button
        exit_button = Button(result_window, text="Exit", command=result_window.destroy, font=('Helvetica', 18), bg="lightblue", relief=RAISED, padx=10, pady=5)
        exit_button.pack(pady=20)
        
    except Exception as e:
        print(f"An error occurred: {e}")

def determine_causes(age, gender, chest_pain, bp, cholesterol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal):
    causes = []
    if chest_pain in [2, 3]:
        causes.append("Chest pain (Non-anginal pain or Asymptomatic)")
    if bp > 140:
        causes.append("High blood pressure")
    if cholesterol > 200:
        causes.append("High cholesterol")
    if fbs == 1:
        causes.append("High fasting blood sugar")
    if restecg in [1, 2]:
        causes.append("Abnormal ECG")
    if thalach < 100:
        causes.append("Low maximum heart rate")
    if exang == 1:
        causes.append("Exercise-induced angina")
    if oldpeak > 2:
        causes.append("High ST depression")
    if slope in [1, 2]:
        causes.append("Abnormal ST slope")
    if ca > 0:
        causes.append("Number of major vessels colored by fluoroscopy greater than 0")
    if thal in [2, 3]:
        causes.append("Fixed or reversible defect in thalassemia")
    
    return "Possible causes: " + ", ".join(causes)

master = Tk()
master.title("Heart Disease Prediction System")

# Define the size of the window
window_width = 700
window_height = 800

# Get the screen dimension
screen_width = master.winfo_screenwidth()
screen_height = master.winfo_screenheight()

# Find the center point
center_x = int(screen_width / 2 - window_width / 2)
center_y = int(screen_height / 2 - window_height / 2)

# Set the position of the window to the center of the screen
master.geometry(f'{window_width}x{window_height}+{center_x}+{center_y}')

# Define font size for labels and entries
font_size = ('Helvetica', 14)

# Create a frame to center the content
frame = Frame(master, bg="lightblue")
frame.pack(pady=10, fill=BOTH, expand=True)

# Add a title label
label = Label(frame, text="Heart Disease Prediction System", fg="black", font=('Helvetica', 24, 'bold'), bg="lightblue")
label.pack(pady=20)

# Create a form frame
form_frame = Frame(frame, bg="lightblue")
form_frame.pack(pady=20, padx=20)

# Create labels and entry widgets inside the form frame
labels = [
    "Enter Your Age", "Select Gender", "Enter Value of Chest Pain", "Enter Value of Resting blood pressure",
    "Enter Value of Cholesterol ", "Enter Value of fasting blood sugar", "Enter Value of restecg", "Enter Value of thalach",
    "Enter Value of Exercise Induced Angina", "Enter Value of oldpeak", "Enter Value of slope", "Enter Value of ca", "Enter Value of Thalassemia"
]

entries = []
dropdown_options = {
    "Chest Pain": ["0 (Normal angina)", "1 (Atypical angina)", "2 (Non-anginal pain)", "3 (Asymptomatic)"],
    "fasting blood sugar": ["0", "1"],
    "restecg": ["0", "1", "2"],
    "Exercise Induced Angina": ["0 (No)", "1 (Yes)"],
    "Thalassemia": ["1 (Normal)", "2 (Fixed defect)", "3 (Reversible defect)"],
    "Gender": ["0 (Female)", "1 (Male)"]
}

# Create variables for dropdown menus
ChestPain_var = StringVar(master)
fastingbloodsugar_var = StringVar(master)
restecg_var = StringVar(master)
ExerciseInducedAngina_var = StringVar(master)
Thalassemia_var = StringVar(master)
gender_var = StringVar(master)

# Set default values
ChestPain_var.set(dropdown_options["Chest Pain"][0])
fastingbloodsugar_var.set(dropdown_options["fasting blood sugar"][0])
restecg_var.set(dropdown_options["restecg"][0])
ExerciseInducedAngina_var.set(dropdown_options["Exercise Induced Angina"][0])
Thalassemia_var.set(dropdown_options["Thalassemia"][0])
gender_var.set(dropdown_options["Gender"][0])  # Default to Female

for i, text in enumerate(labels, start=1):
    Label(form_frame, text=text, font=font_size, bg="lightblue").grid(row=i, column=0, pady=5, sticky=W)
    if text == "Select Gender":
        entry = OptionMenu(form_frame, gender_var, *dropdown_options["Gender"])
    elif text == "Enter Value of Chest Pain":
        entry = OptionMenu(form_frame, ChestPain_var, *dropdown_options["Chest Pain"])
    elif text == "Enter Value of fasting blood sugar":
        entry = OptionMenu(form_frame, fastingbloodsugar_var, *dropdown_options["fasting blood sugar"])
    elif text == "Enter Value of restecg":
        entry = OptionMenu(form_frame, restecg_var, *dropdown_options["restecg"])
    elif text == "Enter Value of Exercise Induced Angina":
        entry = OptionMenu(form_frame, ExerciseInducedAngina_var, *dropdown_options["Exercise Induced Angina"])
    elif text == "Enter Value of Thalassemia":
        entry = OptionMenu(form_frame, Thalassemia_var, *dropdown_options["Thalassemia"])
    else:
        entry = Entry(form_frame, font=font_size)
    if text == "Select Gender":
        entry.grid(row=i, column=1, pady=5, sticky=W, padx=10)
    else:
        entry.grid(row=i, column=1, pady=5, sticky=W)
    entries.append(entry)

e1, e4, e5, e8, e10, e11, e12 = [entries[i] for i in [0, 3, 4, 7, 9, 10, 11]]

# Create the predict button
Button(form_frame, text='Predict', command=show_entry_fields, font=font_size, bg="lightblue").grid(row=len(labels)+1, column=0, columnspan=2, pady=20)

mainloop()
```
